## Задание 2.1

1) Сгенерирйуте данные: рассчитайте значения функций $y = x^3$ и $y = 2*x^3$ на отрезке от -10 до 10 с шагом 1. 

2)Постройте графики функции на одном полотне и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. Убедитесь, что по оси X значения изменяются в диапазоне [-10, 10].

```python
x = range(-10,11)

y = [a**3 for a in x]
y2 = [2*a**3 for a in x]

print x,'\n',y,'\n',y2
```

```python
plt.title(u'Графики двух функций', family='verdana')
plt.xlabel(u'ось Х', family='verdana')
plt.ylabel(u'ось Y', family='verdana')

plt.plot(x,y, linewidth=2.0, label = '$y=x^3$')
plt.plot(x,y2, linewidth=2.0, label = '$y=2*x^3$')
plt.fill_between(x, y, y2, color='g', alpha=0.5, label='place between')
plt.legend(loc = 2)
plt.show()
```

## Задание 2.2

1) Сгенерирйуте данные: рассчитайте значения функций $y = x^2$ и $y = 2*x^2$ на отрезке от -10 до 10 с шагом 1. 

2) Сгенерирйуте данные: рассчитайте значения функций $y = x^3$ и $y = 2*x^3$ на отрезке от -10 до 10 с шагом 1.

На первом изображении постройте графики функции из пункта 1) и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. 
На втором изображении постройте графики функции из пункта 2) и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. 

```python
x = range(-10,11)

y = [a**2 for a in x]
y2 = [2*a**2 for a in x]

print x,'\n',y,'\n',y2

y3 = [a**3 for a in x]
y4 = [2*a**3 for a in x]

print x,'\n',y3,'\n',y4
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.title(u'Графики функций $y=x^2$ и $y=2*x^2$', family='verdana')
plt.xlabel(u'ось Х', family='verdana')
plt.ylabel(u'ось Y', family='verdana')

plt.plot(x,y, linewidth=2.0, label = '$y=x^2$')
plt.plot(x,y2, linewidth=2.0, label = '$y=2*x^2$')
plt.fill_between(x, y, y2, color='c', alpha=0.5, label='place between')
plt.legend(loc = 2)

plt.subplot(122)
plt.title(u'Графики функций $y=x^3$ и $y=2*x^3$', family='verdana')
plt.xlabel(u'ось Х', family='verdana')
plt.ylabel(u'ось Y', family='verdana')

plt.plot(x,y3, linewidth=2.0, label = '$y=x^3$')
plt.plot(x,y4, linewidth=2.0, label = '$y=2*x^3$')
plt.fill_between(x, y3, y4, color='m', alpha=0.5, label='place between')
plt.legend(loc = 2)
plt.show()
```
