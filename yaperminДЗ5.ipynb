```python
# функция для замены кода пассажирских классов (поле Pclass) их названиями (1 = Upper, 2 = Middle, 3 = Lower)
def changePclass(Pclass):
    if Pclass==1:
        return 'Upper'
    elif Pclass==2:
        return 'Middle'
    elif Pclass==3:
        return 'Lower'
```

## Задача 2.4

```python
# функция вовращает значение поля Survived для женщин(1-выжила; 0 - нет)
def returnFemale(row):
    return row['Survived'] if row['Sex']=='female' else 0
```

```python
# суммируем поле Survived для женщин умножаем на 1. (чтобы получить вещественное число) и делим на число общее женщин
sum(df.apply(returnFemale, axis=1))*1./sum((df.Sex=='female'))
```

```python
# функция вовращает значение поля Survived для мужчин(1-выжила; 0 - нет)
def returnMale(row):
    return row['Survived'] if row['Sex']=='male' else 0
```

```python
sum(df.apply(returnMale, axis=1))*1./sum((df.Sex=='male'))
```

## Задание 3.2
Найдите всех пассажиров с аномальной стоимостью билета (поле `Fare`). Будем считать стоимость аномальной, если она отличается от среднего более, чем на 5 стандартных отклонений. Выведите соответствующие строки таблицы на экран.

```python
import numpy as np
```

```python
# функция определения стандартного отклонения для столбца Fare
def returnMeanDeviation(series):
    return np.sqrt(series.apply(lambda x: (x - series.mean())**2).sum()/(series.count()-1))
```

```python
# функция определяет номера строк с аномальной ценой для поля Fare и возвращает этот список
def returnField(series):
    i=0
    lines = list()
    md5 = returnMeanDeviation(series)*5
    for elem in series:
        i+=1
        if (abs(elem - series.mean()) > md5):
            lines.append(i)
    if lines == None:
        lines = []
    return lines
```

```python
df.loc[returnField(df.Fare)]
```

## Задание 3.3
Выведите на экран столбцы, в которых бывают пропущенные значения. (Hint: *Здесь может помочь метод* `.any()`*, а также индексация переменной* `df.columns`).

```python
# функция, которая для каждого столбца из df проверяет, есть ли в нем null значения и если есть, добавляет их в список
def returnFieldNA(df):
    fieldNA = list()
    for elem in df.columns:
        if (df[elem].isnull().any()) == True:
            fieldNA.append(elem) 
    if fieldNA == None:
        fieldNA = []
    return fieldNA
```

```python
df[returnFieldNA(df)]
```
